```
%sql

--DROP TABLE new_table
DROP TABLE IF EXISTS rais_vinculada_2008_2020;
CREATE TABLE rais_vinculada_2008_2020 AS 
(
SELECT ANO, CD_UF, CD_CNAE20_DIVISAO, CD_SEXO, CD_RACA_COR, VL_IDADE, CD_GRAU_INSTRUCAO, CD_TIPO_VINCULO,
  CASE 
    WHEN CD_SEXO = 01 THEN 'H'
    WHEN CD_SEXO = 02 THEN 'M'
   ELSE 'NI'
  END AS CD_SEXO_IDENT,
  CASE 
    WHEN VL_IDADE <= 18 THEN '<= 18'
    WHEN (18 <= VL_IDADE AND VL_IDADE <= 24) THEN '18-24'
    WHEN (25 <= VL_IDADE AND VL_IDADE <= 39) THEN '25-39'
    WHEN (40 <= VL_IDADE AND VL_IDADE <= 59) THEN '40-59'
   ELSE '> ou = 60 '
  END AS FAIXA_ETARIA
FROM df
WHERE FL_VINCULO_ATIVO_3112 = '1'
GROUP BY ANO, CD_UF, CD_CNAE20_DIVISAO, CD_SEXO, CD_RACA_COR, VL_IDADE, CD_GRAU_INSTRUCAO, CD_TIPO_VINCULO
);

--SELECT CD_UF, CD_CNAE20_DIVISAO, CD_SEXO, CD_RACA_COR, VL_IDADE, CD_GRAU_INSTRUCAO, CD_TIPO_VINCULO, CD_SEXO_IDENT, FAIXA_ETARIA FROM rais_vinculada_2008_2020
```
