```python
import pandas as pd
import numpy as np
import json



function_lambda = lambda x: x.replace('<h3>','').replace('</h3>','').replace('\n','').replace('\n','').replace('<p>','').replace('</p>','').replace('<a>','').replace('</a>','').replace('<a','').replace('<p class="',' ').replace('"> href="',' ')

# x = lambda a : a + 10
# print(x(5))
# 15

locais = ['ACRE', 'ALAGOAS','AMAPA','AMAZONAS','BAHIA','CEARA','DISTRITO_FEDERAL',
         'ESPIRITO_SANTO', 'GOIAS', 'MARANHAO', 'MATO_GROSSO', 'MATO_GROSSO_DO_SUL',
         'MINAS_GERAIS', 'PARA', 'PARAIBA', 'PARANA', 'PERNAMBUCO', 'PIAUI', 'RIO_DE_JANEIRO',
         'RIO_GRANDE_DO_NORTE', 'RIO_GRANDE_DO_SUL', 'RONDONIA', 'RORAIMA', 'SANTA_CATARINA',
         'SAO_PAULO', 'SERGIPE', 'TOCANTINS']

lista_de_locais = []
new = []
for local in locais:
    #print(local)
    with open('{local}.txt'.format(local=local),'r') as fobj:
        data = json.load(fobj)
    #data['results'][0:len(data['results'])]
    
    for i in data['results']:
        #print(data['results'][0:len(data['results'])])
        Estado = '{local}'.format(local=local)
        Id_1 = i['id']
        Id_2 = i['uf']['id']
        UF = i['uf']['sigla']
        #Descricao = i['uf']['descricao'] # <----- lambda utilizada
        Descricao = function_lambda(i['uf']['descricao'])
        Unidade = i['nomeUnidade']
        Cidade = i['nomeCidade']
        Bairro = i['nomeBairro']
        Rua = i['nomeRua']
        Endereco = i['numeroEndereco']
        Complemento = i['complemento']
        Telefone = i['telefone']
        Email = i['email']
        Cep = i['cep']
        Lat = i['latitude']
        Long = i['longitude']

# {"count":28,"next":null,"previous":null,"results":[{'id': 777, 'uf': {'id': 1, 'sigla': 'AC'}]
# "results":[{'id': 777, 'uf': {'id': 1, 'sigla': 'AC'}]
# data['results'] = {'id': 777, 'uf': {'id': 1, 'sigla': 'AC'}

# i['numeroEndereco']
# 'numeroEndereco': '2710'
    
        x = {'Estado':Estado,'Id_1': Id_1, 'Id_2':Id_2, 'UF':UF, 'Descricao':Descricao, 'Unidade':Unidade,
             'Cidade':Cidade, 'Bairro':Bairro, 'Rua':Rua, 'Endereco':Endereco, 'Complemento':Complemento,
             'Telefone':Telefone, 'Email':Email, 'Cep':Cep, 'Lat':Lat, 'Long':Long}
        
        lista_de_locais.append(x)
    df =  pd.DataFrame(lista_de_locais)
    #df.to_csv('df_all.csv',index=False)
```
